<h3>Month-wise sales trends and detect seasonal pattern</h3>

```python
mttls=list(Ecom.iloc[:,6:18].sum())
```

```python
mname=list(Ecom.columns)[6:18]
mname
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(14,5))
plt.plot(mname,mttls,marker="*",markersize=12,markerfacecolor="red",color="green")
plt.title("seasonal sales pattern")
plt.xlabel("month name")
plt.ylabel("total sale")
plt.tight_layout()
```

<p>Now we will see top five categories products using pie chart</p>

```python
t5cat=total_sale.sort_values("year",ascending=False).head()
t5cat
```

```python
t5cat.info()
```

```python
plt.figure(figsize=(16,6))
plt.pie(t5cat["year"],labels=t5cat["category"],autopct="%1.1f%%")
plt.title("Top Five Categories Product")
plt.tight_layout()
plt.show()
```
